```python
# import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Kelompok4_DatasetUTS.csv')
df
```

```python
kecamatan = df['Kecamatan']
jumlahKK = df['Jumlah Penerima Bantuan Sosial (Kepala Keluarga)']
x_coords = np.arange(len(kecamatan))
plt.bar(x_coords, jumlahKK)
plt.show()
```

```python
x_coords = np.arange(len(kecamatan))
plt.bar(x_coords, jumlahKK, tick_label=kecamatan)
plt.xticks(rotation=90) #rotates text for x-axis labels
plt.show()
```

```python
x = kecamatan
y = jumlahKK
plt.plot(x, y)
plt.title('Jumlah Penerima Bantuan Sosial (Kepala Keluarga) Menurut Kecamatan di Kabupaten Asahan Tahun 2021')
plt.ylabel('Jumlah Kepala Keluarga')
plt.xlabel('Kecamatan')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(
  kecamatan,
  jumlahKK,
  marker='o',
)
plt.title('Jumlah Penerima Bantuan Sosial (Kepala Keluarga) Menurut Kecamatan di Kabupaten Asahan Tahun 2021')
plt.ylabel('Jumlah Kepala Keluarga')
plt.xlabel('Kecamatan')
plt.xticks(rotation=90)
plt.show()
```
